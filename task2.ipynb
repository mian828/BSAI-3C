```python
def Game(n:int =100):
 for i in range(1,n+1):
        if i % 15==0:
            print("FIZZ BUZZ")
    
        elif i % 3==0:
            print("fizza")
        elif i % 5==0:
            print("Buzz")
        else:
            print(i)
if __name__ == "__main__":
    Game(100)
```

```python
movie_list= [("khan" , 2000000) ,("Game of throne",3000000), ("Sultan" , 1200000), ("Avengers" , 40000000)]
def calculate_average(items):
    total= sum(budget for _, budget in items)
    return total/len (items)
def show_movies (items):
    print("All movies and budget :\n ")
    for name , budget,in items:
        print(f"-{name}: {budget:,.0f}")
        print()

def show_above_aaverage(items ,avg):
    count = 0
    print("movies above the average budget :\n ")
    for name, budget,in items:
        if budget>avg:
            diff= budget-avg
            print(f"{name}:{budget:,.0f}(aboveby{diff:,.0f})")
            count+=1
            print(f"\nTotal movies above average:{count}")

show_movies(movie_list)
avg = calculate_average(movie_list)
print(f"Average budget :{avg:,.0f}\n")
show_above_aaverage(movie_list,avg)
    

    
```
