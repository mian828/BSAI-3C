```python
class Graph:
    def __init__(self, edg_list):
        self.edg_list = edg_list

    def get_neighbors(self, node):
        return self.edg_list[node]

    
    def heuristic(self, node):
        heuristic_values = {
            'A': 1,
            'B': 1,
            'C': 1,
            'D': 1
        }
        return heuristic_values[node]

    def a_star(self, start, goal):
        open_nodes = [start]
        closed_nodes = []

      
        g_score = {start: 0}

        
        parent = {start: start}

        while len(open_nodes) > 0:
            cur_node = None

             
            for node in open_nodes:
                if cur_node is None or g_score[node] + self.heuristic(node) < g_score[cur_node] + self.heuristic(cur_node):
                    cur_node = node

            if cur_node is None:
                print("Path does not exist!")
                return None

            
            if cur_node == goal:
                path = []
                while parent[cur_node] != cur_node:
                    path.append(cur_node)
                    cur_node = parent[cur_node]
                path.append(start)
                path.reverse()

                print("Path found:", path)
                return path

            
            for (N_B, cost) in self.get_neighbors(cur_node):
                if N_B not in open_nodes and N_B not in closed_nodes:
                    open_nodes.append(N_B)
                    parent[N_B] = cur_node
                    g_score[N_B] = g_score[cur_node] + cost
                else:
                    if g_score[N_B] > g_score[cur_node] + cost:
                        g_score[N_B] = g_score[cur_node] + cost
                        parent[N_B] = cur_node

                        if N_B in closed_nodes:
                            closed_nodes.remove(N_B)
                            open_nodes.append(N_B)

            open_nodes.remove(cur_node)
            closed_nodes.append(cur_node)

        print("Path does not exist!")
        return None


 
edg_list = {
    'A': [('B', 1), ('C', 3), ('D', 7)],
    'B': [('D', 5)],
    'C': [('D', 12)]
}

graph = Graph(edg_list)
graph.a_star('A', 'D')
```

```python
class Graph:
    def __init__(self, edges):

        self.edges = edges

    def neighbors(self, node):
        return self.edges.get(node, [])

    def h(self, node):

        h_vals = {
            'A': 1,
            'B': 1,
            'C': 1,
            'D': 1
        }
        return h_vals.get(node, 0)

    def a_star(self, start, goal):
        
        open_list = [start]
        
        closed_list = set()

        
        g_cost = {start: 0}

        
        came_from = {start: None}

        while open_list:
            
            current = min(open_list, key=lambda n: g_cost[n] + self.h(n))

    
            if current == goal:
                path = []
                while current is not None:
                    path.append(current)
                    current = came_from[current]
                path.reverse()
                print("Shortest path:", path)
                return path

            open_list.remove(current)
            closed_list.add(current)

    
            for neighbor, weight in self.neighbors(current):
                if neighbor in closed_list:
                    continue

                tentative_cost = g_cost[current] + weight

                if neighbor not in open_list:
                    open_list.append(neighbor)
                elif tentative_cost >= g_cost.get(neighbor, float('inf')):
                    
                    continue

            
                came_from[neighbor] = current
                g_cost[neighbor] = tentative_cost

        print("No path found.")
        return None


# Example graph
edges = {
    'A': [('B', 1), ('C', 3), ('D', 7)],
    'B': [('D', 5)],
    'C': [('D', 12)]
}

g = Graph(edges)
g.a_star('A', 'D')
```
