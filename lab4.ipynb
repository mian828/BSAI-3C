#LUHN ALOGRITHM

```python

number = input("Enter a number: ")
digits = [int(x) for x in number]  

for i in range(len(digits) - 2, -1, -2):
    doubled = digits[i] * 2
    if doubled > 9:
        doubled = doubled - 9
    digits[i] = doubled

total = sum(digits)
if total % 10 == 0:
    print("Valid number according to Luhn Algorithm.")
else:
    print("Invalid number.")
```

##puncation remove

```python

punctuations = '''!()-[]{};:'" ,<>./?@#$%^&*_~'''

text = input("Enter a string: ")
result = ""

for ch in text:
    if ch not in punctuations:
        result = result + ch

print("this is output:", result)
                                                                                   
```

```python
def arrange_words(text):
    parts = text.split()
    n = len(parts)

    for a in range(n):
        for b in range(0, n - a - 1):
            if parts[b].lower() > parts[b + 1].lower():
                parts[b], parts[b + 1] = parts[b + 1], parts[b]

    return " ".join(parts)
def arrange_letters(text):
    words = text.split()
    new_list = []

    for item in words:
        letters = list(item)
        size = len(letters)

        for x in range(size):
            for y in range(0, size - x - 1):
                if letters[y].lower() > letters[y + 1].lower():
                    letters[y], letters[y + 1] = letters[y + 1], letters[y]

        new_list.append("".join(letters))

    return " ".join(new_list)
line= input("ENter a sentence")
print("Words arranged:", arrange_words(line))
print("Letters arranged:", arrange_letters(line))
```
