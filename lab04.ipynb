```python

def bubble_sort_words(phrase):
    tokens = phrase.split()
    length = len(tokens)
    for step in range(length - 1):
        for pos in range(length - step - 1):
            if tokens[pos].casefold() > tokens[pos + 1].casefold():
                tokens[pos], tokens[pos + 1] = tokens[pos + 1], tokens[pos]
    return " ".join(tokens)


def bubble_sort_letters(phrase):
    tokens = phrase.split()
    arranged = []
    for word in tokens:
        chars = list(word)
        for step in range(len(chars) - 1):
            for pos in range(len(chars) - step - 1):
                if chars[pos].casefold() > chars[pos + 1].casefold():
                    chars[pos], chars[pos + 1] = chars[pos + 1], chars[pos]
        arranged.append("".join(chars))
    return " ".join(arranged)


example_sentence = "grape Orange kiwi Apple"
print("Alphabetical word sort:", bubble_sort_words(example_sentence))
print("Alphabetical letter sort:", bubble_sort_letters(example_sentence))


def filter_text(raw_text):
    output = ""
    for symbol in raw_text:
        if symbol.isalnum() or symbol.isspace():
            output += symbol
    return output


noisy_text = "AI@2025 is#awesome!!"
print("Cleaned string:", filter_text(noisy_text))


card_input = "6011329933655299"
check_digit = int(card_input[-1])
body = card_input[:-1]

print("Card body:", body)
print("Check digit:", check_digit)

rev_body = body[::-1]
digits = [int(x) for x in rev_body]
print("Digits (reversed):", digits)

# doubling rule
for idx, val in enumerate(digits):
    print(f"Index {idx}, original {val}")
    if val % 2 == 0:
        doubled = val * 2
        if doubled > 9:
            doubled -= 9
        digits[idx] = doubled
        print("Changed to:", digits[idx])
    else:
        print("No change (odd number)")

# summation
total = sum(digits)
final_check = total + check_digit
valid = final_check % 10 == 0

print("Sum of digits:", total)
print("Final check value:", final_check)
print("Valid card?", valid)
```
