```python
# Question no : 01
def bfs_without_queue(graph, start):
    visited = set()
    level = {start: 0}   # store level (distance from start)

    def visit_node(node, depth):
        if node not in visited:
            print(node, end=" ")   
            visited.add(node)

        for neighbor in graph.get(node, []):
            if neighbor not in visited:
                level[neighbor] = depth + 1

    depth = 0
    current_level = [start]

    while current_level:
        next_level = []
        for node in current_level:
            visit_node(node, depth)
            for neighbor in graph.get(node, []):
                if neighbor not in visited:
                    next_level.append(neighbor)
        depth += 1
        current_level = next_level


# Example Graph
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}

print("BFS traversal (without queue):")
bfs_without_queue(graph, 'A')
```

```python
# Question no :02
def bfs_with_queue(graph, start):
    visited = []          # list to track visited nodes
    queue = [start]      

    while queue:
        node = queue[0]   
        queue = queue[1:] 

        if node not in visited:
            print(node, end=" ")
            visited.append(node)

            for neighbor in graph[node]:
                if neighbor not in visited and neighbor not in queue:
                    queue.append(neighbor)


# Example Graph
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}

print("BFS traversal (with list as queue):")
bfs_with_queue(graph, 'A')
```
