```python
class Graph:
    def __init__(self, edges):

        self.edges = edges

    def neighbors(self, node):
        return self.edges.get(node, [])

    def h(self, node):

        h_vals = {
            'A': 1,
            'B': 1,
            'C': 1,
            'D': 1
        }
        return h_vals.get(node, 0)

    def a_star(self, start, goal):
        
        open_list = [start]
        
        closed_list = set()

        
        g_cost = {start: 0}

        
        came_from = {start: None}

        while open_list:
            
            current = min(open_list, key=lambda n: g_cost[n] + self.h(n))

    
            if current == goal:
                path = []
                while current is not None:
                    path.append(current)
                    current = came_from[current]
                path.reverse()
                print("Shortest path:", path)
                return path

            open_list.remove(current)
            closed_list.add(current)

    
            for neighbor, weight in self.neighbors(current):
                if neighbor in closed_list:
                    continue

                tentative_cost = g_cost[current] + weight

                if neighbor not in open_list:
                    open_list.append(neighbor)
                elif tentative_cost >= g_cost.get(neighbor, float('inf')):
                    
                    continue

            
                came_from[neighbor] = current
                g_cost[neighbor] = tentative_cost

        print("No path found.")
        return None


# Example graph
edges = {
    'A': [('B', 1), ('C', 3), ('D', 7)],
    'B': [('D', 5)],
    'C': [('D', 12)]
}

g = Graph(edges)
g.a_star('A', 'D')
```
